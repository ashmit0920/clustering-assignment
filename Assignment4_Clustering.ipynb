```python
from pycaret.clustering import *
import pandas as pd

mydataset = pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")
mydataset.head()
```

```python
kMeanClusteringParameters = setup(mydataset)
```

```python
num_clusters=[3,4,5]
sil_values=[]
cal_values=[]
dav_values=[]
```

```python
def find_values(num_clusters):
    model = create_model('kmeans', num_clusters=num_clusters)
    prediction = assign_model(model)
    performance = pull()
    sil_values.append(performance['Silhouette'][0])
    cal_values.append(performance['Calinski-Harabasz'][0])
    dav_values.append(performance['Davies-Bouldin'][0])

def run_clustering():
  for num in num_clusters:
    find_values(num)
```

```python
print("No data processing")
run_clustering()

print("By doing normalization")
normalisation = setup(data=mydataset, normalize=True, verbose=False)
run_clustering()

print("By doing transformation")
transformation = setup(data=mydataset, transformation=True, transformation_method='yeo-johnson', verbose=False)
run_clustering()

print("By doing PCA")
pca = setup(data=mydataset, pca=True, pca_method='linear', verbose=False)
run_clustering()

print("By doing Normalisation + Transformation")
nt = setup(data=mydataset, normalize=True, transformation=True, normalize_method='zscore', transformation_method='yeo-johnson', verbose=False)
run_clustering()

print("By doing Normalisation + PCA")
npca = setup(data=mydataset, normalize=True, pca=True, normalize_method='zscore', pca_method='linear', verbose=False)
run_clustering()

print("By doing Transformation + PCA")
tpca = setup(data=mydataset, transformation=True, pca=True, transformation_method='yeo-johnson', pca_method='linear', verbose=False)
run_clustering()

print("By doing Normalisation + Transformation + PCA")
ntpca = setup(data=mydataset, normalize=True, transformation=True, pca=True, normalize_method='zscore', transformation_method='yeo-johnson', pca_method='linear', verbose=False)
run_clustering()
```
